## Enter replicate measurements
We will use a simple example: 

Imagine you have five cubes of different sizes. You measure the width of each cube three times. In this example, each cube represents your x variable, and the measured widths are your y variable. Each x value has three replicate measurements.

```python
# Dictionary: keys are cube numbers (x-values), values are replicate measurements
replicates = {
    1: [2.18, 2.05, 2.12],
    2: [3.92, 4.08, 3.98],
    3: [6.05, 6.22, 6.10],
    4: [8.16, 8.02, 8.11],
    5: [10.05, 10.22, 10.10]
}

# Create DataFrame, this is a way to store data, similar to a table in excel.
# 'df' is the name of the dataframe, you can change it to anything you want
df = pd.DataFrame.from_dict(
    replicates,
    orient="index",
    columns=["Measurement 1", "Measurement 2", "Measurement 3"]
)

# Name the index
df.index.name = "Cube"
```

```python
# rename the dataframe 
# Rename df to cube
cube = df
```

## Lets create a simple plot

```python
# Example data 
x = [1, 2, 3, 4, 5]
y = [2.1, 4.0, 6.2, 8.1, 10.0]

# Create scatter plot
fig = plt.figure(figsize=(6,6))

plt.scatter(x, y)
plt.xlabel("x values", fontsize = 14)
plt.ylabel("y values", fontsize = 14) 
plt.title("Example scatter plot", fontsize = 16) #plot title

plt.xticks(rotation=45);  # Rotate x-axis
```
