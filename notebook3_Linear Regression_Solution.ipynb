# Notebook 3 — Weighted Linear Regression (Beginner Version)

Goal: Use per-point measurement uncertainties (SEM) to perform a weighted fit of $y=a+bx$, compute parameter uncertainties (standard error of $a$ and $b$) and covariance, and make informative plots (confidence/prediction bands, residuals, covariance ellipse).

## Theory Primer — Why Weighted Fits?

Some points are measured more precisely than others. If each mean value $y_i$ at $x_i$ has an uncertainty $\sigma_i$ (for example, the SEM from Notebook 2), we can give that point a weight $w_i = 1/\sigma_i^2$. 

Assuming that all $x_i$ are measured perfectly (as is approximately the case for a physics experiment, though hardly ever for natural "experiments"), we can then measure if our model

$y=a+bx$ 

works well by computing the $\chi^2$ ("chi-square") misfit for $N$ data

$
\chi^2(a,b) = \sum_{i=1}^N \left(\frac{y_i-a-b x_i}{\sigma_i}\right)^2.
$

This misfit metric uses data weighted by error squared. Other choices are possible, with various benefits and drawbacks. $\chi^2$ metrics have well understood behavior. For example, if $M$ is the number of parameters (here $M=2$, $a$ and $b$), then 

$
\nu = N-M
$

is the number of degrees of freedom, with which we expect $\chi^2 \sim \nu$ for a "moderately" good fit (e.g. Numerical Recipes, sec. 15.1).

We can see to minimize $\chi^2$ by varying $a$ and $b$ any way we like, e.g. by trying out plausible numbers given our prior knowledge about $a$ and $b$ (a "Bayesian" approach), or we can write down the analytical, least squares line fit with errors in $y$ version based on a number of sums (see, e.g., Numerical Recipes, chap. 15.2)

$
S = \Sigma w_i,\quad S_x = \Sigma w_i x_i, \quad S_y = \Sigma w_i y_i,\quad S_{xx} = \Sigma w_i x_i^2,\quad S_{xy} = \Sigma w_i x_i y_i.
$

and then the best fit parameters result as:

$
b = \frac{S S_{xy} - S_x S_y}{\Delta}, a = \frac{S_{xx} S_y - S_x S_{xy}}{\Delta},\quad\mathrm{where}\quad \Delta = S S_{xx} - S_x^2.
$

One can show that the variances (error estimates squared) for the parameters themselves in this case are given by

$
\sigma_a^2 = \frac{S_{xx}}{\Delta}
$

and

$
\sigma_b^2 = \frac{S}{\Delta}.
$

Also important is that those two errors, for intercept, $a$, and slope, $b$, are not independent necessarily, but vary together with the covariance given by 

$
\mathrm{Cov}(a,b) = -\frac{S_x}{\Delta}
$

or the correlation between $a$ and $b$

$
r_{ab} = \frac{-S_x}{\sqrt{SS_{xx}}}
$

if $r_{a,b} \sim 0 $, then both are independently constrained (poorly or well, depending on $\sigma_a$ and $\sigma_b$), if $|r_{a,b}|\sim 1$, then only the ratio of $a/b$ might be constrained. What $\sigma_{a,b}$ and $r_{a,b}$ result from the data is not only determined by the individual errors for $y_i$, $\sigma_i$, but also where those points are sampled (draw a bunch of points to the left and right and centered on $x=0$ and think about this).

## Lets work with data from January to August, where temperature increases each month

```python
jan_aug = data.loc[0:7] #makes a new dataframe of jan - aug
jan_aug = jan_aug.reset_index(drop=True)

jan_aug
```

```python
#convert dataframe to a dictionary
# each month has three measurements of average temperature from 2000, 2010, and 2023
replicates = {
    float(i + 1): [
        row["2000 Average Temp"],
        row["2010 Average Temp"],
        row["2023 Average Temp"]
    ]
    for i, row in jan_aug.iterrows()
}
```

```python
print(replicates)
```

## Compute the Mean and SEM for each month

```python
# Lists to store calculated values
means = []
sems = []

# Loop over each x value and its repeated measurements
for x_value, y_values in replicates.items():
    n = len(y_values)   # number of measurements
    mean_y = sum(y_values) / n  # calculate mean
    std_dev = math.sqrt(sum((y - mean_y)**2 for y in y_values) / (n - 1)) # sample standard dev
    sem = std_dev / math.sqrt(n)  # standard error of the mean

    means.append(mean_y)
    sems.append(sem)

# Convert to NumPy arrays (same as last time)
x = np.array(list(replicates.keys()))
y = means = np.array(means)
sems = np.array(sems)

print('x value:', x)
print('Mean:', y)
print('SEM:', sems)
```

## Weighted fit (by hand) and parameter uncertainties

Now that we estimated the errors for each $y_i$ by computing the mean (for the best estimate) and standard deviation (or variance) for each sample to get $\sigma_i$, we can use the equations for the weighted line fit from the beginning of the exercise. 


$
S = \Sigma w_i,\quad S_x = \Sigma w_i x_i, \quad S_y = \Sigma w_i y_i,\quad S_{xx} = \Sigma w_i x_i^2,\quad S_{xy} = \Sigma w_i x_i y_i.
$

```python
w = 1.0 / (sems**2) #weight of each point

#weighted sums
S = float(np.sum(w)) 
Sx = float(np.sum(w * x))
Sy = float(np.sum(w * y))
Sxx = float(np.sum(w * x**2))
Sxy = float(np.sum(w * x * y))

Delta = S * Sxx - Sx**2

#solve for best fit parameters
b = (S * Sxy - Sx * Sy) / Delta
a = (Sxx * Sy - Sx * Sxy) / Delta
```

### Compute chi-squared, and extract uncertainty in a and b [SE(a) and SE(b)]

After finding the best fit line: $y=a+bx$ 

we can evaluate how well the model describes the data by computing the chi-squared value: 

$
\chi^2(a,b) = \sum_{i=1}^N \left(\frac{y_i-a-b x_i}{\sigma_i}\right)^2.
$

This sums the squared deviations of the data from the fit, weighted by the measurement uncertainties $\sigma_i^2$.

Since we estimate $M$ parameters from the data (here $M = 2$ for $a$ and $b$), the \emph{degrees of freedom} is  

$
\nu = N - M
$

- A `moderately' good fit typically gives $\chi^2 \sim \nu$

To normalize for the number of data points and parameters, we define the reduced chi-squared:

$
\chi^2_\nu = \frac{\chi^2}{\nu}
$

- $\chi^2_\nu \approx 1$ $\Rightarrow$ model describes the data within uncertainties
- $\chi^2_\nu > 1$ $\Rightarrow$ model underestimates scatter or fit is poor
- $\chi^2_\nu < 1$ $\Rightarrow$ uncertainties may be overestimated


The uncertainties in the intercept 𝑎 and slope 𝑏 come directly from the weighted least-squares formulas:

$
\mathrm{SE}(a) = \sqrt{\frac{S_{xx}}{\Delta}}
$

$
\mathrm{SE}(b) = \sqrt{\frac{S}{\Delta}}
$

$
\quad\mathrm{where }\quad S = \Sigma w_i,\quad S_{xx} = \Sigma w_i x_i^2, y_i\quad\Delta = S S_{xx} - S_x^2.
$



```python
# Compute chi-squared to measure goodness of fit
chi2 = np.sum((res / sems)**2) # chi-squared
nu = len(x) - 2  # degrees of freedom (M - number of parameters(a and b))
red_chi2 = chi2 / nu # reduced chi-squared

# Covariance matrix 
Cov = (1.0 / Delta) * np.array([
    [Sxx, -Sx],
    [-Sx,  S]
])

# Scale covariance matrix using reduced chi-squared
Cov_scaled =Cov * red_chi2

# Extract standard errors and covariance
se_a = np.sqrt(Cov_scaled[0, 0]) # uncertainty in intercept a
se_b = np.sqrt(Cov_scaled[1, 1]) # uncertainty in slope b
cov_ab = Cov_scaled[0, 1] # covariance between a and b

# Print results
print("SE(a) =", se_a)
print("SE(b) =", se_b)
print("Cov(a, b) =", cov_ab)
print("chi^2 =", chi2)
print("reduced chi^2 =", red_chi2)
```

### Compute Confidence and Prediction Intervals

When we fit a line 𝑦=𝑎+𝑏𝑥, we often want to quantify uncertainty in our predictions. There are two common intervals:

1. Confidence interval (CI)
    - Gives the range where we expect the true mean value of $y$ to lie at a given $𝑥_0$. 
    - Narrower than the prediction interval because it only reflects uncertainty in the fit parameters.

    - $
y_{\text{fit}} \pm t^* \, \mathrm{SE}_{\text{mean}}(x_0)
$


1. Prediction interval (PI)
    - Gives the range where a new observation is likely to fall at $𝑥_0$. 
    - Always wider than CI bevause it includes both fit uncertainty and measurement scatter.

    - $
y_{\text{fit}} \pm t^* \, \mathrm{SE}_{\text{pred}}(x_0)
$


$
\mathrm{SE}_{\mathrm{mean}}(x_0) = \sqrt{\mathbf{v}^T \, \mathrm{Cov} \, \mathbf{v}}, 
\quad 
\mathbf{v} = 
\begin{bmatrix} 1 \\ x_0 \end{bmatrix}
$


$
\mathrm{SE}_{\mathrm{pred}}(x_0) = \sqrt{ \mathrm{SE}_{\mathrm{mean}}(x_0)^2 + \sigma_{\mathrm{new}}^2 }
$


Notes:
- t* = t-critical value (≈1.96 for 95% confidence in N>30)
- If the sample size is small (N<30) the table of t-distribution accounts for extra uncertainty
- CI → uncertainty in the mean fitted line
- PI → uncertainty for a new measurement

```python
# 95% t critical values table for small sample sizes
t95 = {1:12.706, 2:4.303, 3:3.182, 4:2.776, 5:2.571,
       6:2.447, 7:2.365, 8:2.306, 9:2.262, 10:2.228,
       15:2.131, 20:2.086, 25:2.060, 30:2.042}

df = len(x) - 2                 # degrees of freedom
t_star = t95.get(df, 1.96)      # default to 1.96 if df not in table

# Standard error of the mean prediction at x0
def se_mean(x0, Cov):
    v = np.array([1.0, x0])
    return np.sqrt(v @ Cov @ v)

# Standard error for a new predicted observation at x0
sigma_typical = np.mean(sems)  # typical measurement scatter
def se_pred(x0, Cov):
    return np.sqrt(se_mean(x0, Cov)**2 + sigma_typical**2)

# Compute SEs across grid of x values
se_mean_grid = np.array([se_mean(x0, Cov_scaled) for x0 in x_grid])
se_pred_grid = np.array([se_pred(x0, Cov_scaled) for x0 in x_grid])

# Compute 95% confidence and prediction intervals
y_ci_low  = y_fit - t_star * se_mean_grid
y_ci_high = y_fit + t_star * se_mean_grid

y_pi_low  = y_fit - t_star * se_pred_grid
y_pi_high = y_fit + t_star * se_pred_grid
```

## Covariance Ellipse
Another useful plot to visualize uncertainty in our fit is a covariance ellipse. 

In the linear fit 𝑦 = 𝑎 + 𝑏𝑥, the covariance ellipse shows the joint uncertainty of the intercept and slope. 

The width of the ellipse along the a-axis reflects the uncertainty in the intercept.
The height along the b-axis reflects the uncertainty in the slope.

The tilt of the ellipse shows the correlation: if the ellipse is slanted downward, a larger intercept tends to correspond to a smaller slope, which is typical in regression.

All points inside the ellipse represent combinations of (a,b) that are statistically consistent with the data and the weighted linear fit. The ellipse gives a visual sense of how precisely the line is determined and how the slope and intercept uncertainties are related.

```python
# Compute the "directions and lengths" of uncertainty for the parameters
# Eigenvectors give the main directions in which a and b vary together
# Eigenvalues give how big the uncertainty is along each direction
vals, vecs = np.linalg.eigh(Cov_scaled)

# Make a circle of points (unit circle) to start with
theta = np.linspace(0, 2*np.pi, 200)
circle = np.stack([np.cos(theta), np.sin(theta)])  # 2D circle points

# Stretch and rotate the circle to make an ellipse using the covariance info
# The ellipse shows the combined uncertainty of a (intercept) and b (slope)
ellipse = (vecs @ np.diag(np.sqrt(vals)) @ circle).T

# Plot the ellipse and the best-fit point
plt.figure(figsize=(6,5))
plt.plot(a + ellipse[:,0], b + ellipse[:,1], 'b-')  # ellipse in blue
plt.plot([a], [b], 'ro')   # best-fit parameters in red
plt.xlabel('a (intercept)')
plt.ylabel('b (slope)')
plt.title('Parameter Covariance Ellipse (scaled)')
plt.axis('equal')   # equal scaling for x and y
plt.grid(True, alpha=0.3)
```

## Recompute standard error and Chi-squared

```python
#Define a function to recompute specified values, see below.
def recompute(x, y, sems):
    """
    Recompute weighted linear fit parameters, residuals, chi-squared, 
    and parameter uncertainties.

    Inputs: x : array, y : array-like, sems: standard errors

    Returns: results in a dict
        - a : intercept
        - b : slope
        - res : residuals
        - chi2 : chi-squared
        - red_chi2 : reduced chi-squared
        - se_a : uncertainty in a
        - se_b : uncertainty in b
        - cov_ab : covariance between a and b
        - Cov_scaled : scaled covariance matrix
    """
    
    # Compute weights (smaller errors → higher weight)
    w = 1.0 / (sems**2)

    # Weighted sums
    S   = np.sum(w)
    Sx  = np.sum(w * x)
    Sy  = np.sum(w * y)
    Sxx = np.sum(w * x**2)
    Sxy = np.sum(w * x * y)

    # Determinant for solving linear equations
    Delta = S * Sxx - Sx**2

    # Solve for weighted best-fit slope (b) and intercept (a)
    b = (S * Sxy - Sx * Sy) / Delta
    a = (Sxx * Sy - Sx * Sxy) / Delta

    # Compute residuals
    res = y - (a + b * x)

    # Chi-squared and reduced chi-squared
    chi2 = np.sum((res / sems)**2)
    nu = len(x) - 2  # degrees of freedom (N - number of parameters)
    red_chi2 = chi2 / nu

    # Covariance matrix of parameters
    Cov = (1.0 / Delta) * np.array([[Sxx, -Sx],
                                    [-Sx,  S]])
    # Scale covariance matrix by reduced chi-squared
    Cov_scaled = Cov * red_chi2

    # Standard errors and covariance
    se_a = np.sqrt(Cov_scaled[0, 0])
    se_b = np.sqrt(Cov_scaled[1, 1])
    cov_ab = Cov_scaled[0, 1]

    # Print results
    print("Slope (b) =", b)
    print("Intercept (a) =", a)
    print("SE(a) =", se_a)
    print("SE(b) =", se_b)
    print("Cov(a, b) =", cov_ab)
    print("Chi-squared =", chi2)
    print("Reduced chi-squared =", red_chi2)

    # Return everything in a dictionary for further use
    results = {
        'a': a,
        'b': b,
        'res': res,
        'chi2': chi2,
        'red_chi2': red_chi2,
        'se_a': se_a,
        'se_b': se_b,
        'cov_ab': cov_ab,
        'Cov_scaled': Cov_scaled
    }

    return results
```
